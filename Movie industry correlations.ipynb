```python
# Importing the usgae packages for this project.

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Reading in the data.
df = pd.read_csv(r'C:\Users\jeffb\Downloads\movies.csv')
```

```python
df.head()
```

```python
df = df.dropna()
```

```python
# Identify missing data.

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print ('{} - {}%'.format(col, pct_missing))
```

```python
# printing dtypes

df.dtypes
```

```python
df['budget'] = df['budget'].fillna(0).astype(np.int64)
print (df['budget'])

df['gross'] = df['gross'].fillna(0).astype(np.int64)
print (df['gross'])
```

```python
df.dtypes
```

```python
# Create correct year column

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df
```

```python
# Sorting data

df = df.sort_values (by = ['gross'], inplace=False, ascending=False)
```

```python
# Display all data

pd.set_option('display.max_rows', None)
```

```python
# Dropping duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Bidget for Film')

plt.show()
```

```python
df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matric for Numeric features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
correlation_matt = df_numerized.corr()

corr_pairs = correlation_matt.unstack()

corr_pairs



```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```
